```python
import pandas as pd
import matplotlib.pyplot as plt
dataframe=pd.DataFrame({
    "car":['BMW','Lexus','Audi','Mustang','Bentey','Jaguar'],
    "Reg_Price":[2000,2500,2800,3000,3200,3500],
    "Units":[100,120,150,170,180,200]
    })
plt.plot(dataframe["Reg_Price"],
         dataframe["Units"])
plt.xlbel("Registration Price")
plt.ylabel("Units sold")
plt.title("Units sold vs Registration price")
plt.show()
```

```python
import numpy as np
x=np.array([[1,2],[4,5]])
y=np.array([[7,8],[9,10]])
print("The element wise addition of matrices is: ")
print(np.add(x,y))
print("The element wise subtraction of matrices is: ")
print(np.subtract(x,y))
print("The element wise division of matrices is: ")
print(np.divide(x,y))
print("The element wise multiplication of matrices is: ")
print(np.multiply(x,y))
print("The matrix product of matrices is: ")
print(np.dot(x,y))
```
